#subir daraset desde un archivo

```
import pandas as pd
#read a locakl csv file named 'data.scv'
df= pd.read_csv('clase.csv')
df_aux = df
#display the first few row of the dataFrame
print(df.head())
```

1. Convertir todas las columnas en minusculas

```
df.columns = df.columns.str.lower()
display(df.head())
```

2. Eliminar todos los acentos de las palabras

```
import unicodedata
import re

def remove_accents_and_special_chars(text):
    if isinstance(text, str):
        # Normalize to NFD (Canonical Decomposition Form) and remove combining characters
        text = unicodedata.normalize('NFD', text)
        text = ''.join(c for c in text if unicodedata.category(c) != 'Mn')
        # Remove non-alphanumeric characters except spaces and hyphens
        text = re.sub(r'[^a-zA-Z0-9\s-]', '', text)
        return text
    return text

# Apply the function to all string columns in the DataFrame
for col in df.columns:
    if df[col].dtype == 'object': # Check if the column contains strings
        df[col] = df[col].apply(remove_accents_and_special_chars)

display(df.head())
```

3. Ekiminar los caracteres especiales

```
import re

def remove_all_special_chars(text):
    if isinstance(text, str):
        # Keep only alphanumeric characters
        text = re.sub(r'[^a-zA-Z0-9]', '', text)
        return text
    return text

# Apply the function to all string columns in the DataFrame
for col in df.columns:
    if df[col].dtype == 'object': # Check if the column contains strings
        df[col] = df[col].apply(remove_all_special_chars)

display(df.head())
```

4. Contar cuantos valores nulos tiene

```
print(df.isnull().sum())
```

5. Contar cuantos valores vacios tiene

```
empty_counts = {}
for col in df.columns:
    if df[col].dtype == 'object':
        empty_counts[col] = (df[col] == '').sum()

print("Contador de valores vacios (empty strings) por columna:")
for col, count in empty_counts.items():
    print(f"{col}: {count}")
```
